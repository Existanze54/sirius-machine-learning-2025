```python
df = pd.read_csv('data/gse53625_expression.csv', index_col=0)
mdf = pd.read_csv('data/gse53625_metadata.csv', index_col=0)
```

```python
df.head(2)
```

```python
mdf.head(2)
```

```python
X = df.T
mark = mdf['Sample type'] == 'cancer'
```

Обучим и применим PCA.

```python
pca = PCA(n_components=2)
drX = pca.fit_transform(X)
```

Визуализируем.

```python
x = drX[:, 0]
y = drX[:, 1]

plt.scatter(x[~mark], y[~mark])
plt.scatter(x[mark], y[mark])

plt.legend(['normal', 'cancer'])
plt.show()
```
