#### Задача

Нарисуйте на одном холсте график $\sin x$, $\cos x$ и $\dfrac{\cos 2x}{2}$ на промежутке от $-\pi$ до $\pi$. Отобразите:
1. График $\sin x$ - жирной зеленой пунктирной линией.
2. График $\cos x$ - полужирной оранжевой штриховой линией.
3. График $\frac{\cos 2x}{2}$ - тонкой фиолетовой штрихпунктирной линией.

```python
x = np.linspace(-np.pi, np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.cos(2*x) / 2

plt.plot(x,y1, lw=5 ,color='green', ls="--")
plt.plot(x,y2, lw=2 ,color='orange', ls=":")
plt.plot(x,y3, lw=1 ,color='violet', ls="-.")
```

#### Задача

Скопируйте код из предыдущей задачи и добавьте к нему легенду.

```python
x = np.linspace(-np.pi, np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.cos(2*x) / 2

plt.plot(x,y1, lw=5 ,color='green', ls="--", label="$\sin x$")
plt.plot(x,y2, lw=2 ,color='orange', ls=":", label="$\cos x$")
plt.plot(x,y3, lw=1 ,color='violet', ls="-.", label="$\dfrac{\cos 2x}{2}$")

plt.legend(loc="center left")
```

### Графики разных типов в MPL все строятся примерно так же!

Вам обычно достаточно немного поменять функцию, и вы получите другой график!

### Совместное распределение двух переменных

```python
x = np.random.normal(0, 1, 100)
y = (x + np.random.normal(0, 1, 100)) / 2
```

```python
plt.scatter(x, y, color="k", marker="x")
```

```python
m, b = np.polyfit(x, y, 1)  # Regression
x0, x1 = x.min(), x.max()
y0, y1 = m * x0 + b, m * x1 + b
plt.scatter(x, y, color="grey", marker="x")
plt.plot([x0, x1], [y0, y1], color="black", ls="--")
print(f"y={m:.2f}x{b:+.2f}")
```

### Какие графики есть в Seaborn?

### `sns.relplot`

**Базовый scatter plot**:

```python
heart_uci.head()
```

```python
sns.relplot(x="Age", y="Max_heart_rate", data=heart_uci)
```

Можно показывать третье измерение цветом:

```python
sns.relplot(x="Age",
            y="Max_heart_rate",
            hue="Exercise_angina",
            data=heart_uci)
```

...или размером точек:

```python
sns.relplot(x="Age",
            y="Max_heart_rate",
            size="Vessels_colored",
            data=heart_uci)
```

**Не вносите** слишком много информации на один график, иначе он получится **перегруженным**!

```python
sns.relplot(x="Age",
            y="Max_heart_rate",
            hue="Chest_pain_type",
            size="Vessels_colored",
            style="Sex",
            data=heart_uci)
```

Вместо этого лучше разносить это на разные графики!

```python
sns.relplot(x="Age",
            y="Max_heart_rate",
            size="Vessels_colored",
            col="Chest_pain_type",
            row="Sex",
            data=heart_uci,
            height=3.5)
```

### Графики для распределений:
`displot`, `catplot` (`boxplot`, `violinplot`, `stripplot`, `swarmplot`)

Можно посмотреть на распределения величины с помощью `sns.displot`:

```python
sns.displot(x="Age", data=heart_uci)
```

Также можно отобразить эмпирическую плотность:

```python
sns.displot(x="Age", kde=True, data=heart_uci)
```

Можно раскрашивать в **группировке**:

```python
sns.displot(x="Age",
            hue="Sex",
            kde=True,
            data=heart_uci)
```

```python
sns.displot(x="Age", hue="Sex", data=heart_uci, alpha=0.5)
```

```python
sns.displot(x="Age", hue="Sex", data=heart_uci, alpha=0.5, element="step")
```

Но с большим числом групп уже некрасиво...
